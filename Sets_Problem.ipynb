```python
s = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
results = []

def find_zero_sum(current_list, start_index):
    # If current list has 3â€“6 elements, and sums to 0, save a copy
    if 3 <= len(current_list) <= 6 and sum(current_list) == 0:
        results.append(current_list[:])  # append a copy

    # Stop if already 6 elements
    if len(current_list) == 6:
        return

    # Try adding next elements one by one
    for i in range(start_index, len(s)):
        current_list.append(s[i])
        find_zero_sum(current_list, i + 1)  # move to next index (no repeats)
        current_list.pop()  # backtrack

# Start with empty list from index 0
find_zero_sum([], 0)

# Print results
for r in results:
    print(r)

print(f"\nTotal such lists found: {len(results)}")
```
